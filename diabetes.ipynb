# Addressing the Class Imbalance Problem: Train-Test Split

```python
#Class Imbalance

zero, one = (y.value_counts())
print("Diabetes to No Diabetes ratio: "+str(one/zero))
data = [zero,one]
lab = ('0: No diabetes','1: (Pre)Diabetes')
plt.figure()
plt.title("Bar plot of positive and negative diabetes diagnoses in y")
plt.ylabel("Number of instances")
plt.bar(lab,data)
plt.show()
print("Figure i: Illustrating the class imbalance in our dataset using a bar plot")
```

According to google devs, $\approx 18\%$ minority to majority class ratio is considered of moderate imbalance. This means we need to employ a downsampling or oversampling principle.
